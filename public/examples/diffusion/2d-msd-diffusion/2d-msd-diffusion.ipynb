# Calculation of diffusion coefficient using mean square displacement
We can estimate the diffusion coefficient using the [mean square displacement](https://en.wikipedia.org/wiki/Mean_squared_displacement#Derivation_of_the_MSD_for_a_Brownian_particle_in_1D). The mean square displacement is defined as

$MSD = \langle (\vec x(t) - \vec x_0)^2\rangle  = \frac{1}{N}\sum_{i=1}^{N} |\vec x^{(i)}(t) - \vec x^{(i)}_0|^2$

This quantity relates to the diffusion coefficient through (when $t\rightarrow \infty$)

$\langle (\vec x(t) - \vec x_0)^2\rangle = 2Dt$

where $D=2$ is the number of dimensions.

In this simulation, we measure the mean square displacement using [compute msd](https://docs.lammps.org/compute_msd.html), and calculate the diffusion coefficient

$D = \frac{\langle (\vec x(t) - \vec x_0)^2\rangle}{2D}$

```python
import lammps_logfile
import matplotlib.pyplot as plt

log = lammps_logfile.File("log.lammps")
step = log.get("Step")
plt.figure()

plt.plot(step, log.get("c_msd[4]"))
plt.xlabel('Time')
plt.ylabel('Mean square displacement')
plt.title('Mean square displacement')
plt.show()

plt.figure()
plt.plot(step, log.get("v_diffusioncoeff"))
plt.xlabel('Time')
plt.ylabel('Diffusion coefficient')
plt.title('Diffusion coefficient')

plt.show()
```
